```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
from cycler import cycler
rcParams['figure.figsize'] = 18, 5
rcParams['axes.spines.top'] = False
rcParams['axes.spines.right'] = False
rcParams['axes.prop_cycle'] = cycler(color=['#365977'])
rcParams['lines.linewidth'] = 2.5
from sklearn.utils import resample
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
import plotly.express as px 
from sklearn.metrics import confusion_matrix, plot_confusion_matrix,\
    precision_score, recall_score, accuracy_score, f1_score, log_loss,\
    roc_curve, roc_auc_score, classification_report
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from fbprophet import Prophet as proph
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
import holidays
holiday = pd.DataFrame([])
for date, name in sorted(holidays.UnitedStates(years=[2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]).items()):
    holiday = holiday.append(pd.DataFrame({'ds': date, 'holiday': "US-Holidays"}, index=[0]), ignore_index=True)
holiday['ds'] = pd.to_datetime(holiday['ds'], format='%Y-%m-%d', errors='ignore')
```

# GOLD 

```python
df_gold = pd.read_csv('gold.csv', index_col='Date', parse_dates=True)
df_gold
```

```python
test_size = 300


df_gold_train = df_gold[:-test_size]
df_gold_test = df_gold[-test_size:]

plt.title('Gold train and test sets', size=20)
plt.plot(df_gold_train, label='Training set')
plt.plot(df_gold_test, label='Test set', color='orange')
plt.legend();
```

```python
df_gold_train
```

```python
df_gold_test
```

```python
df_gold_test.reset_index(inplace=True)
df_gold_train.reset_index(inplace=True)
```

```python
df_gold_test
```

```python
df_gold_train
```

```python
# Rename the columns [Date,Close] to [ds, y]

ts_gold = df_gold_train

ts_gold = df_gold_train.rename(columns={'Date': 'ds',
                        'Close': 'y'})

ts_gold_test = df_gold_test.rename(columns={'Date': 'ds',
                        'Close': 'y'})
ts_gold.tail(5)
```

```python
x_gold_test = ts_gold_test[['ds']]
```

```python
ts_gold_test
```

```python
# set the uncertainty interval to 95% (the Prophet default is 80%)

Model_gold = proph(interval_width=0.95,daily_seasonality=True,holidays=holiday)
```

```python
# Fit the timeseries into Model
Model_gold.fit(ts_gold)
```

```python
# Use make_future_dataframe with a daily frequency and periods = 1518 for 6 years
future_gold_dates = Model_gold.make_future_dataframe(periods=1518, freq='D')
future_gold_dates.tail()
```

```python
future_gold_dates= future_gold_dates.loc[future_gold_dates['ds'] >= '2020-11-30']
```

```python
x_gold_test = pd.DataFrame(x_gold_test)
```

```python
x_gold_test
```

```python
# Predict the values for future dates and take the head of forecast

predicitons_gold = Model_gold.predict(x_gold_test)
```

```python
x_gold_test
```

```python
forecast_gold = Model_gold.predict(future_gold_dates)
```

```python
future_gold_dates.head(15)
```

```python
forecast_gold.tail()
```

```python
gold_roi = forecast_gold.copy()
gold_roi.set_index('ds',inplace=True)
gold_last_value = pd.DataFrame()
gold_last_value['last_value_month'] = gold_roi['yhat'].resample('M').apply(lambda ser: ser.iloc[-1,])
gold_last_value['ROI'] = gold_last_value['last_value_month'] / ts_gold_test.iloc[0,1]*100 - 100
```

```python
gold_roi.index
```

```python
gold_roi.info()
```

```python
#gold_roi['yhat'].resample('M').apply(lambda ser: ser.iloc[-1,])
```

```python
gold_last_value
```

```python
gold_last_value.rename(columns={'ROI':'Gold_ROI'}, inplace=True)
```

```python
Model_gold.plot(forecast_gold, uncertainty=True)
plt.show()
```

```python
df_gold_test = pd.concat([df_gold_test,predicitons_gold], axis=1)
df_gold_test
```

```python
predicitons_gold
```

```python
df_gold_test.yhat.value_counts()
```

```python
mean_squared_error(df_gold_test.Close,df_gold_test.yhat,squared=False)
```

## Plotting Forecast vs Actuals

```python
f, ax = plt.subplots(figsize=(14,5))
f.set_figheight(5)
f.set_figwidth(15)
ts_gold_test.plot(kind='line',x='ds', y='y', color='red', label='Actual', ax=ax)
predicitons_gold.plot(kind='line',x='ds',y='yhat', color='green',label='Forecast', ax=ax)
plt.title('Gold Forecast vs Actuals')
plt.show()
```

```python
x_gold_test
```

### MAPE

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
mape = mean_absolute_percentage_error(ts_gold_test['y'],predicitons_gold['yhat'])
print("MAPE",round(mape,4))
```

# Tunning

```python
from sklearn.model_selection import ParameterGrid
params_grid = {'seasonality_mode':('multiplicative','additive'),
               'changepoint_prior_scale':[0.1,0.2,0.3],     
              'n_changepoints' : [100,150,200]}
grid = ParameterGrid(params_grid)
cnt = 0
for p in grid:
    cnt = cnt+1

print('Total Possible Models',cnt)
```

```python
model_parameters = pd.DataFrame(columns = ['MAPE','Parameters'])
for p in grid:
    test = pd.DataFrame()
    print(p)
    train_model =proph(changepoint_prior_scale = p['changepoint_prior_scale'],
                         n_changepoints = p['n_changepoints'],
                         seasonality_mode = p['seasonality_mode'],
                         weekly_seasonality=True,
                         daily_seasonality = True,
                         yearly_seasonality = True, 
                         interval_width=0.95)
    
    train_model.fit(ts_gold)
    train_forecast = x_gold_test
    train_forecast = train_model.predict(train_forecast)
    test=train_forecast[['ds','yhat']]
    Actual = ts_gold_test
    MAPE = mean_absolute_percentage_error(Actual['y'],abs(test['yhat']))
    print('Mean Absolute Percentage Error(MAPE)------------------------------------',MAPE)
    model_parameters = model_parameters.append({'MAPE':MAPE,'Parameters':p},ignore_index=True)
```
